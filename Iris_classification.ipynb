```python
X=df.iloc[:,[0,1,2]]
```

```python
X
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X=scaler.fit_transform(X)    
```

```python
X
```

```python
import matplotlib.pyplot as plt
```

```python
fig=plt.figure(figsize=(10,5))
from mpl_toolkits.mplot3d import Axes3D
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0], X[:,1],X[:,2], s=20,c='purple',label='Species')
plt.title='Species before clustering'
plt.show()
```

```python
plt.scatter(X[:, 0], X[:, 1], s=20, c='green', label='Species')
```

```python
from sklearn.cluster import KMeans
wccs=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X)
    wccs.append(kmeans.inertia_)
plt.plot(range(1, 11), wccs)
plt.show()

```

```python
kmeans=KMeans(n_clusters=3,random_state=42)
K_means=kmeans.fit_predict(X)
```

```python
K_means
```

```python
kmeans.cluster_centers_
```

```python
df['K_means']=K_means
```

```python
df.tail(60)
```

```python
plt.scatter(X[K_means==0,0],X[K_means==0,1],c='b')
plt.scatter(X[K_means==1,0],X[K_means==1,1],c='r')
plt.scatter(X[K_means==2,0],X[K_means==2,1],c='g')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],c='purple')
plt.show()
```

```python
 #just for comparing results(this is not requirement in task) in K_means and real 'Species' columns we need to change Spaces to numerical data and find accuracy_score
df.loc[df[df['Species']=='Iris-setosa'].index, 'Species']=1 
```

```python
df.loc[df[df['Species']=='Iris-versicolor'].index, 'Species']=2
```

```python
df.loc[df[df['Species']=='Iris-virginica'].index, 'Species']=0
```

```python
df['Species']=df['Species'].astype('int32')
```

```python
df.tail(20)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score=accuracy_score(df['Species'],df['K_means']) 
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
dendrogram = dendrogram(linkage(X, method='ward'))
plt.title=('Dendrogram:Hierarchcal Clustering')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hierarchical_c=AgglomerativeClustering(n_clusters=3)
y_hc=hierarchical_c.fit_predict(X)
```

```python
y_hc
```

```python
plt.scatter(X[y_hc==0,0], X[y_hc==0,1], c='black')
plt.scatter(X[y_hc==1,0], X[y_hc==1,1], c='red')
plt.scatter(X[y_hc==2,0], X[y_hc==2,1], c='green')
plt.show()
```

```python
df['H_Clustering']=y_hc
```
